```python
from pyspark.sql.types import (
    StructType,
    StructField,
    IntegerType,
    StringType,
    DateType,
    LongType
)

from pyspark.sql.functions import current_timestamp
```

```python
spark.sql("""
    CREATE VOLUME IF NOT EXISTS paola_bronze.default.bronze_f1
""")
```

```python
raw_path = "/Volumes/paola_bronze/default/rawf1"

drivers_file = f"{raw_path}/drivers.csv"
races_file = f"{raw_path}/races.csv"
lap_times_file = f"{raw_path}/lap_times.csv"

bronze_path = "/Volumes/paola_bronze/default/bronze_f1"
```

```python
display(dbutils.fs.ls(raw_path))
```

```python
drivers_schema = StructType([
    StructField("driverId", IntegerType(), False),
    StructField("driverRef", StringType(), True),
    StructField("number", IntegerType(), True),
    StructField("code", StringType(), True),
    StructField("forename", StringType(), True),
    StructField("surname", StringType(), True),
    StructField("dob", DateType(), True),
    StructField("nationality", StringType(), True),
    StructField("url", StringType(), True)
])
```

```python
drivers_df = (
    spark.read
        .option("header", True)
        .option("nullValue", "\\N")
        .option("dateFormat", "yyyy-MM-dd")
        .schema(drivers_schema)
        .csv(drivers_file)
        .withColumnRenamed("driverId", "driver_id")
        .withColumnRenamed("driverRef", "driver_ref")
        .withColumn("ingestion_timestamp", current_timestamp())
)

drivers_df.printSchema()
display(drivers_df.limit(10))
```

```python
races_schema = StructType([
    StructField("raceId", IntegerType(), False),
    StructField("year", IntegerType(), True),
    StructField("round", IntegerType(), True),
    StructField("circuitId", IntegerType(), True),
    StructField("name", StringType(), True),
    StructField("date", DateType(), True),
    StructField("time", StringType(), True),
    StructField("url", StringType(), True),
    StructField("fp1_date", DateType(), True),
    StructField("fp1_time", StringType(), True),
    StructField("fp2_date", DateType(), True),
    StructField("fp2_time", StringType(), True),
    StructField("fp3_date", DateType(), True),
    StructField("fp3_time", StringType(), True),
    StructField("quali_date", DateType(), True),
    StructField("quali_time", StringType(), True),
    StructField("sprint_date", DateType(), True),
    StructField("sprint_time", StringType(), True)
])
```

```python
races_df = (
    spark.read
        .option("header", True)
        .option("nullValue", "\\N")
        .option("dateFormat", "yyyy-MM-dd")
        .schema(races_schema)
        .csv(races_file)
        .withColumnRenamed("raceId", "race_id")
        .withColumnRenamed("circuitId", "circuit_id")
        .withColumn("ingestion_timestamp", current_timestamp())
)

races_df.printSchema()
display(races_df.limit(10))
```

```python
lap_times_schema = StructType([
    StructField("raceId", IntegerType(), False),
    StructField("driverId", IntegerType(), False),
    StructField("lap", IntegerType(), False),
    StructField("position", IntegerType(), True),
    StructField("time", StringType(), True),
    StructField("milliseconds", LongType(), True)
])
```

```python
lap_times_df = (
    spark.read
        .option("header", True)
        .option("nullValue", "\\N")
        .schema(lap_times_schema)
        .csv(lap_times_file)
        .withColumnRenamed("raceId", "race_id")
        .withColumnRenamed("driverId", "driver_id")
        .withColumn("ingestion_timestamp", current_timestamp())
)

lap_times_df.printSchema()
display(lap_times_df.limit(10))
```

```python
(
    drivers_df.write
        .format("delta")
        .mode("overwrite")
        .save(f"{bronze_path}/drivers")
)

(
    races_df.write
        .format("delta")
        .mode("overwrite")
        .save(f"{bronze_path}/races")
)

(
    lap_times_df.write
        .format("delta")
        .mode("overwrite")
        .save(f"{bronze_path}/lap_times")
)
```

```python
drivers_bronze_df = spark.read.format("delta").load(
    f"{bronze_path}/drivers"
)

races_bronze_df = spark.read.format("delta").load(
    f"{bronze_path}/races"
)

lap_times_bronze_df = spark.read.format("delta").load(
    f"{bronze_path}/lap_times"
)

print(f"Drivers: {drivers_bronze_df.count()} registros")
print(f"Corridas: {races_bronze_df.count()} registros")
print(f"Voltas: {lap_times_bronze_df.count()} registros")
```
